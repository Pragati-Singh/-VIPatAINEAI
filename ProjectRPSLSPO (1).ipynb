# **Rock - Paper - Scissors - Lizard - Spock**

```python
import random                                                                          # importing the library
```

Rock_paper_scissors_lizard_spock.png

**r = Rock, p = Paper, s = Scissors, l = Lizard, spo = Spock**

```python
begin = 'y'                                                                           
while begin.lower() == 'y':
   moves = ['r','p','s', 'l', 'spo']                                                                           # possible moves
   player_wins = ['pr','sp','rs', 'lp', 'lspo', 'spos','spor', 'rl', 'sl','pspo']                              # winning possibility
   player_moves = input('Enter your move "r = Rock, p = Paper, s = Scissors, l = Lizard, spo = Spock:"')      # user input
   computer_moves = random.choice(moves)                                                                       # random computer moves from random.choices
   print("Computer's move is {} :".format('computer_moves'))                                                   # printing through string formatting method
   print("Player's move is: %s" %(player_moves))                                                               # printing through string formatting method       
    
   if player_moves == computer_moves:                                                                          # Implementing NestedIf loop  
       print("tie")
   elif player_moves + computer_moves in player_wins:
       print("you win")
   else: 
       print("you lose")
  
   begin = input("Do you want to continue? y = yes, q = quit : ")                      # taking user input in the variable begin that whether to continue the game or not
   if begin.lower() == 'q':                                                            # converting the input to lower case and implenting if and break. 
       break
```
